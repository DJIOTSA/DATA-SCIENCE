Perform exploratory data analysis on the netflix_data.csv data to understand more about movies from the 1990s decade.

What was the most frequent movie duration in the 1990s? Save an approximate answer as an integer called duration (use 1990 as the decade's start year).

A movie is considered short if it is less than 90 minutes. Count the number of short action movies released in the 1990s and save this integer as short_movie_count.

```python
# Importing pandas and matplotlib
%pip install pandas numpy matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")
```

```python
# Start coding here! Use as many cells as you like
print(netflix_df.columns)

# get only the 1990's movies
all_1990s_movies = netflix_df[(netflix_df['type']== 'Movie') &( netflix_df['release_year'] >= 1990) & (netflix_df['release_year'] <= 1999)]

# get just th value of the duration collum
duration_series = all_1990s_movies.loc[:, ['release_year','duration']]

print(duration_series)

# let visualize the duraion
plt.figure(figsize=(10, 5))
plt.plot(duration_series['release_year'], duration_series['duration'], marker='o', linestyle='-')

plt.xlabel('Release Year')
plt.ylabel('Duration (minutes)')
plt.title('Movie Duration in the 1990s')
plt.grid(True)
plt.show()

# the most frequent movie duration
duration = all_1990s_movies['duration'].mode()[0]
print(' the most frequent movie duration in the 1990s: ' + str(duration))

#  short_movie_count duration < 90
short_movies = all_1990s_movies[(all_1990s_movies['duration'] < 90) & (all_1990s_movies['genre'] == 'Action')]

short_movie_count = short_movies['duration'].size

print(short_movie_count)

```
