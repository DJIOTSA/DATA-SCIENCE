# Changing data

```python
#Adding a 'Total' column of all the stats added together 
pokemon_data_csv['Totals']= pokemon_data_csv['HP'] + pokemon_data_csv['Attack'] + pokemon_data_csv['Defense'] + pokemon_data_csv['Sp. Atk'] + pokemon_data_csv['Sp. Def'] + pokemon_data_csv['Speed']
print(pokemon_data_csv)
```

```python
# Removing a column
# Hence to remove it, you need to create another variable to remove/drop that column
new_pokemon_data_csv = pokemon_data_csv.drop(columns=['Totals'])
print(new_pokemon_data_csv.head())
```

```python
# Summing multiple columns and making a new column
# In iloc[0:10, 4:10], 
# the first 0:10 indicates you are selecting the indexes of the rows you want to have data in this new column, placing only : 
# indicates you want all rows
# the second 4:10 indicates you are selecting the indexes of the columns
 
# .sum(axis=1) means you want to add the rows together as values in the new column 'Total'
# .sum(axis=0) means you want to add the columns together as values in the new column Total'
pokemon_data_csv['Totals'] = pokemon_data_csv.iloc[0:10, 4:10].sum(axis=1)
print(pokemon_data_csv)
```

# Regex Filtering

```python
import re
#.str.contains() is able to find a certain strings from rows in the
#columns and print rows with that string out
print(pokemon_data_csv[pokemon_data_csv['Name'].str.contains('Mega')].head(2))
#To take rows that does not have that certain string you add the wiggly line 
print(pokemon_data_csv[~pokemon_data_csv['Name'].str.contains('Mega')])
```

```python
new_data2 = pokemon_data_csv.loc[(pokemon_data_csv['Type 1'] == 'Grass') & (pokemon_data_csv['Type 2'] == 'Poison') & (pokemon_data_csv['HP'] > 70)]
print(new_data2)
print('----------------------')
#Recall this,code, and that we can also use these RegEx functions to filter out data similarly
#Flag variable is used as a signal in programming to let the program know that a certain condition has met. 
#It usually acts as a boolean variable indicating a condition to be either true or false

#re.I, the I stands for 'IGNORECASE', ignore upper and lower case of input (grass in this case) 
#while searching the input through the columns
#The regex Function applies a regular expression to a string and returns the matching substrings
new_data3 = pokemon_data_csv.loc[(pokemon_data_csv['Type 1'].str.contains('grass', flags=re.I, regex=True)) & (pokemon_data_csv['Type 2'].str.contains('Poison', flags=re.I, regex=True)) & (pokemon_data_csv['HP'] > 70) ]
print(new_data3)
#re.I, the I stands for 'IGNORECASE', ignore upper and lower case of input (fire or grass in this case) 
new_data4 = pokemon_data_csv.loc[pokemon_data_csv['Type 1'].str.contains('fire|grass', flags=re.I, regex=True)]
print(new_data4)
```

```python
# ^ means starting with eg: ^pi (starting with pi)
# [A-Z] means character can be any of A to Z
# [a-z]* means zero or more characters behind the first 2 letters 'pi' can be any character from letter a to z 
d = pokemon_data_csv.loc[pokemon_data_csv['Name'].str.contains('^pi[a-z]*', flags=re.I, regex=True)]
print(d)
```

# Conditional change

```python
#How to convert all the boxes in column 'Type 1' with the word, 'Fire' to 'Flamer' instead
pokemon_data_csv.loc[pokemon_data_csv['Type 1']=='Fire', 'Type 1'] = 'Flamer'
print(pokemon_data_csv)

#To change 'Flamer' back to 'Fire'
pokemon_data_csv.loc[pokemon_data_csv['Type 1'] == 'Flamer', 'Type 1'] = 'Fire'
print(pokemon_data_csv)
```

```python
#   To make all 'Fire' type pokemon's 'Legendary' status be True
#   Show that you can use condition from 1 column to set the parameter of another column 
#   and edit the other column
pokemon_data_csv.loc[pokemon_data_csv['Type 1'] == 'Fire', 'Legendary'] = True
pokemon_data_csv[pokemon_data_csv['Type 1']=='Fire']
```

```python
#If 'Total' column has value higher than 500, set values for those rows(pokemon) for the 'Generation' 
# column to 'Test 1' and 'Legendary' column to 'Test 2'
pokemon_data_csv = pd.read_csv('pokemon_data.csv')
pokemon_data_csv['Totals'] = pokemon_data_csv.iloc[:, 4:10].sum(axis=1)
# rearrange
columns = list(pokemon_data_csv.columns)
pokemon_data_csv = pokemon_data_csv[columns[0:4] + [columns[-1]] + columns[4:len(columns)-1]]
# perform the operation
pokemon_data_csv.loc[pokemon_data_csv['Totals'] > 500, ['Generation','Legendary']]= ['Test 1', 'Test 2']
pokemon_data_csv
```

```python
# want to modify the original dataframe (pokemon_data_csv) directly:
# Generation = Defense * 3
# Legendary = Atk / 2
# Only for rows where Totals > 500.

# Method 1
pokemon_data_csv.loc[pokemon_data_csv['Totals'] > 500, ['Generation']]= (
    pokemon_data_csv.loc[pokemon_data_csv['Totals'] > 500, 'Defense']*3)
pokemon_data_csv.loc[pokemon_data_csv['Totals']>500, 'Legendary'] = (
    pokemon_data_csv.loc[pokemon_data_csv['Totals']>500, 'Attack']/2)

pokemon_data_csv
```

```python
# Method 2
pokemon_data_csv.loc[pokemon_data_csv['Totals'] < 300, ['Generation','Legendary']]= (
    pokemon_data_csv.loc[pokemon_data_csv['Totals'] < 300].assign(
        Generation = lambda x: x['Defense']*3,
        Legendary = lambda x: x['Attack']/2
    )
)
pokemon_data_csv[['Totals','Generation','Legendary']] 
```

# GroupBy

```python
#Most used groupby functions are .mean(), .sum(), .count()
a = (pokemon_data_csv.drop(columns=['#'])
     .groupby(['Attack'])
     .mean(numeric_only=True)
     .reset_index()).sort_values('Totals', ascending=False)
a
```

```python
a = (pokemon_data_csv.drop(columns=['#'])
     .groupby(['Defense'])
     .sum(numeric_only=True)
     .reset_index()).sort_values('Totals', ascending=False)
a
```

```python
a = (pokemon_data_csv.drop(columns=['#'])
     .groupby(['Generation', 'HP'])
     .count()
     ).sort_values('Totals')
a
```

# Analyze large data with chunk

```python
# with read_csv

# empty dataframe
df = pd.DataFrame(columns=pokemon_data_csv.columns)
for chunk in pd.read_csv('new_text.txt', chunksize=5):
    result = chunk.groupby(['Name'], as_index=False).sum()
    df = pd.concat([df, result])

df
```

```python
# excel
excel_data = pd.read_excel('new_excel.xlsx')
chunksize =  5
for i in range(0, len(excel_data), chunksize):
    chunk = excel_data.iloc[i:i+chunksize]
    print(chunk)
```
