# Mathplotlib

```python
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd
```

```python
# initial data
dataset = pd.read_csv('Automobile_data.csv')
filtered = dataset.groupby(['company'], as_index=False).sum(numeric_only=True)
y= filtered['price']
z = filtered['horsepower']
w=filtered['average-mileage']
x = filtered['company']
filtered
```

# Basics

##### Plotting a graph with '.plot(x,y)'. Usually takes in 1D arrays representing the x and y coordinates in the x and y axis. 
##### plot(x, y): plot x and y using default line style and color
##### plot(x, y, 'bo'): plot x and y using blue circle markers
##### plot(y): plot y using x as index array 0..N-1
##### plot(y, 'r+'): ditto, but with red plusses
##### If x and/or y are 2D arrays a separate data set will be drawn for every column. 
##### If both x and y are 2D, they must have the same shape. If only one of them is 2D with shape (N, m) the other must have length N and will be used for every data set m.

```python
plt.figure(figsize=(20,9))
plt.plot(x,y, scalex=True)
plt.title('My diagram')
plt.xlabel('Companies')
plt.ylabel('Prices')
plt.show()
```

```python
plt.figure(figsize=(20,9))
plt.plot(x,y,'bo')
plt.title('My diagram')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```

```python
plt.plot(w,)
plt.title('My diagram')
plt.ylabel('Cars Prices')
plt.show()
```

```python
plt.plot(z, 'r+')
plt.title('My diagram')
plt.xlabel('X axis')
plt.ylabel('z axis')
plt.show()
```

# changing font style (title, axis,...)

```python
#You can check in the documentation on all the marker (type), linestyle, (line)color supported by matplotlib
plt.figure(figsize=(20,9))
plt.plot(x, z, label='company car horsepower', color='blue', linewidth=3, marker='.', markersize=30, markeredgecolor='blue')
plt.plot(x,w, 'r^--', label='company car length') # shortcut styling

# Adding a labels and title
plt.title('Changing font style', fontdict={
    'fontname': 'Montserrat',
    'fontsize': 20, 
    'fontweight': 'bold'
    })
plt.xlabel('Companies name', fontdict={
    'fontname': 'Arial',
    'color': 'blue'
})
plt.ylabel('cars prices', fontdict={
    'fontname': 'Comic Sans MS',
    'color': 'blue'
})
# changing tick mark
# plt.xticks([1000,2809,3000,])
# plt.yticks([])

# adding legend
plt.legend()



plt.savefig('my-grpah.svg')

# display the graph
plt.show()
```

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

#First graph
plt.plot(x,y, 'r^-', label='y = 2x')

#Second graph
x2 = np.arange(0,4.5,0.5)
#plt.plot(x2, x2**2, 'b^-', label='y = x^2')

#You could do something cool with the second graph like making the first half of the graph a solid line,
#then second half as dotted lines
plt.plot(x2[:6], x2[:6]**2, 'b^-', label='y = x^2')    #(up till index 6)
plt.plot(x2[5:], x2[5:]**2, 'b^--')   #made the indexing overlap abit so the graph looks better (index 5 onwards)

plt.title('Our First Graph!', fontdict={'fontname':'Comic Sans MS', 'fontsize':20, 'fontweight':'bold'})

plt.xlabel('x axis')
plt.ylabel('y axis')

plt.xticks([0,1,2,3,4])
#plt.yticks([0,2,4,6,8])  #Commented this out so that the ticks for y will stretch to all the way to top of y axis

plt.legend()

#Saving the graph
#You can also specify the pixels for your graph that you want to save (while the one you see
#in the code may have a different pixel size)
plt.savefig('linegraph3.png', dpi=1000)

plt.show()

```
