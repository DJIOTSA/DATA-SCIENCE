# Matplotlib Pratique

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('company_sales_data.csv')
months = dataset['month_number']
month_list = list(months)
dataset
```

# Read Total profit of all months and show it using a line plot

```python
total_profits = dataset['total_profit']

plt.plot(months, total_profits, label='Total profits per month', color='green', marker='.', markersize=10)
plt.title('Total profits per month', fontdict={
    'color':'Orange',
    'fontweight': 'bold'
})
plt.xlabel('Months')
plt.ylabel('Total month profits')
plt.xticks(month_list)
plt.legend(loc='lower right')
plt.show()
```

```python
plt.figure(figsize=(5,3))

plt.plot(dataset['month_number'], dataset['total_profit'],  'r.--', label='Profit data of last year', linewidth=3, markersize=20, markerfacecolor='black')

plt.xlabel('Month number')
plt.ylabel('Sold units in number')
plt.title('Company profit per month')

plt.yticks([100000, 200000, 300000, 400000, 500000])
plt.xticks(dataset['month_number'])

plt.legend(loc='lower right')

plt.savefig('prac2_bar_graph.png', dpi=100)

plt.show()
```

# Read all product sales data and show it using a multiline plot


```python
colors = ['blue', 'yellow', 'green', 'red', 'orange', 'teal', 'navy']
i = 0
plt.figure(figsize=(16,9))
for column, data in dataset.items():
    if (column != 'month_number') & (column != 'total_profit') & (column != 'total_units'):
        plt.plot(months, data, 'r.-', label=column, color=colors[i], markersize=10)
        i=i+1

plt.xlabel('Month number')
plt.ylabel('Sold units in number')
plt.title('Company profit per month')
plt.xticks(dataset['month_number'])

plt.legend(loc='upper left')

plt.show()
```

#  Read toothpaste sales data of each month and show it using a scatter plot

```python
plt.figure(figsize=(5,4))
print(dataset.head(2))
plt.scatter(months, dataset['toothpaste'], linewidth=3, color='black', label='Toothpaste sale per month')
plt.title('Toothpaste sales', fontdict={
    'color': 'purple',
    'fontweight': 'bold',
})
plt.xlabel('months')
plt.ylabel('sales')
plt.xticks(month_list)
plt.grid(True, linestyle='--', linewidth=1)
plt.legend(loc='upper left')
plt.show()
```

# Read face cream and facewash product sales data and show it using the bar chart

```python
plt.figure(figsize=(9,5))
plt.bar(months - 0.2,dataset['facecream'], color='pink', width=0.4, label='Face Cream')
plt.bar(months + 0.2,dataset['facewash'], color='blue', width=0.4, label='Face Wash')
plt.xlabel('Month numbers')
plt.ylabel('Sold units in number')
plt.xticks(month_list)
plt.grid(True, linestyle='--', linewidth=0.5)
plt.title('Face Cream and Face Wash sales data')

plt.legend(loc='upper left')
plt.show()
```

# Read the total profit of each month and show it using the histogram to see the most common profit ranges

```python
plt.figure(figsize=(9,5))
bins = [150000, 175000, 200000, 225000, 250000, 300000, 350000]
plt.hist(dataset['total_profit'], bins=bins, label='Total profit')
plt.xlabel('profit range in $')
plt.ylabel('Number of months')
plt.xticks(bins)
plt.grid(True, linestyle='--', linewidth=0.5)
plt.title('Profit')
plt.legend(loc='upper left')
plt.savefig('bathing_soap_bar_graph.png', dpi=100)

plt.show()
```

# Calculate total sale data for last year for each product and show it using a Pie chart

```python
#  subsetting products sales
total_sale_products = dataset.iloc[:, 1:7]
# transpose th data such columns becomes rows
total_sale_products_trans = total_sale_products.T.copy()
total_sale_products_trans['total_sale'] = total_sale_products_trans.iloc[:, :].sum(axis=1)
total_sale_units = total_sale_products_trans['total_sale']
products = total_sale_products.columns
total_sale_units

explode = []
for i in products:
    explode.append(0.1)

plt.pie(total_sale_units, labels=products, autopct='%.1f %%', explode=explode, shadow=True)

#'bbox_to_anchor=(1, 0.5)' function allows you to shift the position of the legend in the graph. The first
#parameter shifts it at the xaxis, while second parameter shifts it at the yaxis
plt.legend(bbox_to_anchor=(0.1, 0.6))

plt.show()
```
